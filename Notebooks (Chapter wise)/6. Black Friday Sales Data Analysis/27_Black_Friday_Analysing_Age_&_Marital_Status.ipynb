```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

# **Analyzing Age**

# **Q1. Total No of purchases from each group ?**

```python
df.groupby('Age').size()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Age Group vs Total No of Purchases')
```

```python
df.groupby('Age').size().plot(kind = 'pie', autopct = '%.2f%%', figsize = (9, 9), title = 'Age Group vs Total No of Purchases')
```

# **Q2. Total amount spend by every age group ?**

```python
df.groupby('Age')['Purchase'].sum()
```

```python
df.groupby('Age')['Purchase'].sum().plot(kind = 'bar', figsize = (12, 6), title = 'Age Group vs Total Amount Spend')
```

```python
df.groupby('Age')['Purchase'].sum().plot(kind = 'pie', autopct = '%.2f%%', figsize = (9, 9), title = 'Age Group vs Total Amount Spend')
```

# **Q3. From each age group, how much avg price they're paying ?**

```python
df.groupby('Age')['Purchase'].mean()
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'bar', figsize = (12, 6), title = 'Age Group vs Avg Price Paying')
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'pie', autopct = '%.2f%%', figsize = (9, 9), title = 'Age Group vs Avg Price Paying')
```

# **Q4. How many unique no of products people are purchasing from each age group ?**

```python
df.groupby('Age')['Product_ID'].nunique()
```

```python
df.groupby('Age')['Product_ID'].nunique().plot(kind = 'bar', figsize = (12, 6), title = 'Age Group vs No of Unique Purchases')
```

```python
df.groupby('Age')['Product_ID'].nunique().plot(kind = 'pie', autopct = '%.2f%%' ,figsize = (9, 9), title = 'Age Group vs No of Unique Purchases')
```

# **Q5. How many males and females are married & unmarried ?**

```python
df.groupby('Marital_Status').size()     # 0 -> unmarried, 1 -> married
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct = '%.2f%%', figsize = (7, 7), title = 'Marital Status')
```
