# **Importing necessery libraries**

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
import statistics as st
```

# **removing all the null values before proceeding further**

```python
df = pd.read_csv('googleplaystore.csv')
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(df.iloc[:, 2:3].values)
df.iloc[:, 2:3] = imputer.transform(df.iloc[:, 2:3])
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

# **Q. Name the apps that belong to the art and design category and also with the rating more than 4.5?**

# **Previously we did using in built functions...**

```python
df_pr = df[(df['Category'] == 'ART_AND_DESIGN') & (df['Rating'] > 4.5)]
df_pr['App']
```

# **GroupBy concept in pandas**

# **Q1. Name the apps that belong to the ART_AND_DESIGN category with rating more than 4.5 in ascending order with respect to rating?**

```python
df_pr = df[(df['Category'] == 'ART_AND_DESIGN') & (df['Rating'] > 4.5)]
```

# **dataframe can be sorted using sort_values() function where we need to paas two parameters: on which column we want to sort & we want to sort it in ascending or descending order**

```python
df_pr.sort_values(by = 'Rating', ascending = True)
```

# **For descending order just make ascending = False. If you want to display the top 10 apps only just add the head() function and pass 10**

```python
df_pr.sort_values(by = 'Rating', ascending = False).head(10)
```

# **Q2. Which category is having the maximum average rating with descending order?**

```python
df_pr = df.groupby('Category')['Rating'].mean().sort_values(ascending = False)
df_pr
```

```python
df_pr.idxmax()  # prints the category having maximum avg rating
```

```python
df_pr.idxmin()  # prints the category having maximum avg rating
```
