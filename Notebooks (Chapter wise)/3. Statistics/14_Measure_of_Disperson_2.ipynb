```python
import statistics as st
```

# **Relative Measure of Dispersion**

# Coefficient of Variation
# Coefficient of Quartile Deviation

# **a. Coefficient of Variation -> Obtained by dividing the std deviation by mean and expressed in percentage**

```python
data1 = [45, 42, 42, 41, 40]
data2 = [54, 48, 42, 33, 33]
```

```python
cov1 = (st.pstdev(data1)/st.mean(data1))*100
cov2 = (st.pstdev(data2)/st.mean(data2))*100
```

```python
cov1
```

```python
cov2
```

# **From this we can say data1 is more stable bcz the value of coefficient of variance of cov1 is less than cov2**

```python
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
(st.pstdev(df['mpg'])/st.mean(df['mpg'])) * 100  # cov of mpg column
```

```python
(st.pstdev(df['weight'])/st.mean(df['weight'])) * 100  # cov of weight column
```

# **weight column is more stable than mpg column bcz the cov value is lesser in case of weight column**

# **b. Coefficient of Quartile Deviation -> (Q3 - Q1)/(Q3 + Q1)**

```python
df.describe()
```

```python
Q3 = 29.0          #for mpg col
Q1 = 17.5
cqv = (Q3 - Q1)/(Q3 + Q1)
cqv
```

```python
Q3 = 3608.0          # for weight col
Q1 = 2223.75
cqv = (Q3 - Q1)/(Q3 + Q1)
cqv
```

# **(as cqv of weight) < (cqv of mpg col), means the values of weight col are more stable as compared to the mpg**
