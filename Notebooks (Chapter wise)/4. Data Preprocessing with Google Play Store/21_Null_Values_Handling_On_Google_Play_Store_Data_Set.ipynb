# **Importing numpy and pandas**

```python
import numpy as np
import pandas as pd
```

# **reading the google play store data set**

```python
df = pd.read_csv('googleplaystore.csv')
df
```

# **This time we're gonna follow the imputer approach to avoid any data loss due to the null values**

# **checking in which columns null values are present**

```python
df.isnull().sum()
```

# **null values are present in Rating, Type, Content Rating, Current Ver & Android Ver columns**

# **Importing & defining the imputer**

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

# **at first dealing with the Rating column because 1474 rows are there!**

```python
impute.fit(df.iloc[:, 2:3].values)
df.iloc[:, 2:3] = impute.transform(df.iloc[:, 2:3].values)
df.head()
```

# **as we can see the Rating column contains 0 null values**

```python
df.isnull().sum()
```

```python
df
```

# **when you use .fit() and .transform() methods with SimpleImputer, and print the whole dataframe, the imputed values are stored as floating-point numbers in the DataFrame. For df.head() function it's the default case, no floating nos**

```python
df.isnull().sum()
```

# **After handling the Rating col, there's hardly 13 rows that contain null values. So we can easily remove those columns which will keep the 99.9% data**

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

# **so now the whole data set is free of null values**
# **we removed 12 rows from it, but that's pretty minor as compared to 10841 rows initially availabe to us**
