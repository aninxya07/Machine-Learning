```python
clean1 = df[df['Hour']<6]
print("Cleaning Data from 12:00 am to 5:59 am")
clean1.head(10)

#Cleaning Data from 12:00 am to 5:59 am
```

```python
df = df.drop(clean1.index, axis = 0)
df = df.drop(clean2.index, axis = 0)
```
