# **Importing necessercy libraries**

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
df.head()
```

```python
df.isnull().sum()
```

# **removing all the null values before proceeding further**

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df.iloc[:, 2:3].values)
df.iloc[:, 2:3] = imputer.transform(df.iloc[:, 2:3])
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

```python
for i in df['Category'].unique():
  print(i)
```

# **Handling multiple columns using loops (own logic)**

# **Q1. How many free apps are there in the art and design category?**

# **dataframe is converted to a numpy array and now it can be accessed by passing an index which will give a row from the dataframe**

```python
df_array = df.to_numpy()
df_array[0]
```

```python
count = 0
for i in df_array:
  if (i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
    count += 1
print("No of free apps in Art And Design category are", count)
```

# **Q2. How many paid apps are there in the dating category?**

```python
count = 0
for i in df_array:
  if (i[1] == 'DATING' and i[6] == 'Paid'):
    count += 1
print("No of free apps in Dating category are", count)
```

# **Q3. How many apps are there in the Art And Design category with rating more than 4.5**

```python
count = 0
for i in df_array:
  if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5):
    count += 1
print("No of apps in Art And Design category with rating more than 4.5 are", count)
```

# **Q4. How many apps are there in the Family category with rating more than 4.5 and it's free as well**

```python
count = 0
for i in df_array:
  if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5 and i[6] == 'Free'):
    count += 1
print("No of apps in Art And Design category with rating more than 4.5 and it's also free are", count)
```
