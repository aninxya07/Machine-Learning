# **Checking if there is any null values in the dataset...**
# **there can be thousands of rows in a dataset, so it's obvious we're not manually going to check for null values...**

```python
data.isnull()
```

# **where there is no null value, it'll show false, otherwise true**
# **but it's like the same thing as manual check, we've to traverse through the whole data set in order to check for the null value (finding where true is written)**

```python
data.isnull().sum()
```

# **we're calculating the sum of no of true (null values) present inside each of the columns**
# **looking at these, there're 1474 null values inside Rating column, as well as 1, 1, 8, 3 null values present inside Type, Content Rating, Current Ver, Android Ver**

# **deleting all the rows that contain null values**

```python
data = data.dropna()
data
```
