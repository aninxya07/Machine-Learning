# **Importing numoy and pandas**

```python
import numpy as np
import pandas as pd
```

# **reading a data set using pandas**

```python
df = pd.read_csv('Data.csv')
df
```

# **importing & defining imputer**

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
```

# **applying on age & salary col**
# **for handling numeric values we can follow the strategy of mean, median also...most frequent strategy is mostly used in case of strings**

```python
imputer.fit(df.iloc[:, 1:3].values)
df.iloc[:, 1:3] = imputer.transform(df.iloc[:, 1:3].values)
df
```

# **handling country, purchased using most frequent strategy**

```python
imputer.fit(df.iloc[:, 0:1].values)
df.iloc[:, 0:1] = imputer.transform(df.iloc[:, 0:1].values)
df.iloc[:, 0:1]
```

```python
imputer.fit(df.iloc[:, 3:].values)
df.iloc[:, 3:] = imputer.transform(df.iloc[:, 3:].values)
df.iloc[:, 3:]
```

# **now printing the whole data set and now there's no null values**

```python
df
```
