```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=5, n_features=2, random_state=3)
```

```python
X       # 500 x 2 matrix
```

```python
y       # contains information about each points that belong to which cluster from 0 to 4
```

```python
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1], c = y)   # color scheme
plt.show()
```

```python
k = 5
colors = ['green', 'yellow', 'blue', 'cyan', 'red']
```

```python
clusters = {}
for idx in range(k):
  center = 10*(2*np.random.random((X.shape[1],))-1)
  points = []
  cluster = {
      'center' : center,
      'points' : [],
      'color' : colors[idx]
  }
  clusters[idx] = cluster
clusters
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
for i in clusters:
  center = clusters[i]['center']
  plt.scatter(center[0], center[1], marker = '^', c = 'red')
plt.show()
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
v1 = np.array([1,2])
v2 = np.array([3,4])
distance(v1,v2)
```

```python
#Implementing E step
def assign_clusters():
    for idx in range(X.shape[0]):
        dist = []

        curr_x = X[idx]

        for i in range(k):
            dis = distance(curr_x,clusters[i]['center'])
            dist.append(dis)
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)



#Implementing the M-Step
def update_clusters():
    for idx in range(k):
        pts = np.array(clusters[idx]['points'])
        if pts.shape[0] > 0:
            new_center = pts.mean(axis =0)
            clusters[idx]['center'] = new_center

            clusters[idx]['points'] = []

def plot_clusters():
    for i in clusters:
        pts = np.array(clusters[i]['points'])
        try :
            plt.scatter(pts[:,0],pts[:,1],c=clusters[i]['color'])
        except:
            pass
        center = clusters[i]['center']
        plt.scatter(center[0],center[1],c = 'black',marker = '^')
```

```python
assign_clusters()
plot_clusters()
update_clusters()   # although we achieved the centers but the centers aren't properly placed till now
                    # it requires further optimization which needs lots of mathematics, so we will not
                    # do all the computations manually each & every time, rather we will use the
                    # scikit library for this
```
