```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=5, n_features=2, random_state=3)
```

```python
X       # 500 x 2 matrix
```

```python
y       # contains information about each points that belong to which cluster from 0 to 4
```

```python
from sklearn.cluster import KMeans   # importing the library
```

```python
km = KMeans(n_clusters = 5)     # setting the no of clusters as 5
```

```python
km.fit(X,y)    # calculates the cluster centers in the dataset
```

```python
centers = km.cluster_centers_     # function to print the centers
centers
```

```python
labels = km.labels_     # function to print every points belong to which cluster
labels
```

```python
plt.scatter(X[:,0], X[:,1], c = labels)
plt.scatter(centers[:,0], centers[:,1], c = 'black', marker = '^')      # gives the optimized value
plt.show()
```
