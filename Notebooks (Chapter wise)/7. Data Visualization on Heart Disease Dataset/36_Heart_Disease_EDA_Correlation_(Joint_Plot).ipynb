```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix)
```

# **Joint plot**

```python
sns.jointplot(x = 'Age', y = 'Age', data = df)   # straight line bcz of perfect correlation = 1
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df)      # correlation = -0.38
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df)    # correlation = 0.40
```

```python
sns.jointplot(x = 'HeartDisease', y = 'RestingBP', data = df)    # almost 0 correlation (0.1)
```

```python
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df)
```

```python
sns.jointplot(x = 'Cholesterol', y = 'Oldpeak', data = df)    # correlation = 0.05 almost zero correlation
```

# **creating different visualizations in joint plot**

# **1. Hexagon**

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex')
```

# **2. Regression line**

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'reg')  # shows how the MaxHR is changing based upon age
                                                                # by the regression line

# If the age increases, the MaxHR will decreses. The line tells us this info.
```

```python
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df, kind = 'reg')

# If the person is not having any heart disease then, the maxHR is more (density lies near 140 & 160)
# If the person is having a heart disease, then the maxHR is less (density lies near 120 & 140)
```
