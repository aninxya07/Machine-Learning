```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df['Age'].unique()
```

```python
sns.displot(df['Age'])   # 90% of the data lies between age 30 to 75
```

```python
sns.displot(df['Age'], kde = True)    # draws a line over the distribution
```

# **displot() and distplot() both are different. By default kde is always True in distplot()**

```python
sns.distplot(df['Age'], kde=False, hist = True)     # draws a histogram
```

```python
sns.distplot(df['Age'], kde=False, hist = False, rug = True)     # draws a density distribution
```

```python
sns.distplot(df['Age'], kde = True, hist = True, rug = True)     # all of them combined
```

```python
sns.distplot(df['Age'], kde = False, bins = 10)   # divides the whole histogram into 10 different segments
```

```python
sns.distplot(df['Age'], kde = True, rug = True, bins = 10)
```
