# **Imporrting numpy and pandas**

```python
import numpy as np
import pandas as pd
```

# **reading the data set using pandas**

```python
df = pd.read_csv('Data.csv')
df
```

# **1. Removing the rows**

```python
df.dropna()    #previous approach which is not efficient
```

# **2. Imputer**

# **importing simple imputer from impute function of sklearn library**

```python
from sklearn.impute import SimpleImputer
```

# **defining the imputer on the basis of most frequent function and on null(nan) values**

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
```

# **extracting only the age & salary col on which operation is going to be performed**

```python
res = df.iloc[:, 1:3]
res
```

# **extracting the values**
# **4.4 x 10^1 = 44.0000000 like this values are stored**

```python
res = res.values
res
```

# **passing the required columns into the imputer function in order to perform the operations & then transforming it**
# **then storing it back into a varaible to see the transformed values**

```python
imputer.fit(res)
x = imputer.transform(res)   #at the 6th index the new value is 2.7... x 10^1 = 27.0
x
```

# **putting those modified values back into the columns**

```python
df.iloc[:, 1:3] = x
df.iloc[:, 1:3]
```
