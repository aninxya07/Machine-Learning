```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('BlackFriday.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

# **Analysing single columns using seaborn**

# **Q1. How many customers are there from each city?**

```python
sns.set(rc = {'figure.figsize' : (6, 6)})
sns.countplot(x = 'City_Category', data = df)
```

# **Analysing multiple columns using seaborn library**

# **Q1. Showing age groups on the basis of gender**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'Age', hue = 'Gender', data = df)
```

# **Q2. Showing gender on the basis of age groups**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'Gender', hue = 'Age', data = df)
```

# **Q3. Showing age groups on the basis of marital status**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'Age', hue = 'Marital_Status', data = df)
```

# **Q4. Showing marital status on the basis of age groups**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'Marital_Status', hue = 'Age', data = df)
```

# **Q5. Showing gender on the basis of marital status**

```python
sns.set(rc = {'figure.figsize' : (8, 8)})
sns.countplot(x = 'Gender', hue = 'Marital_Status', data = df)
```

# **Q6. Showing marital status on the basis of gender**

```python
sns.set(rc = {'figure.figsize' : (8, 8)})
sns.countplot(x = 'Marital_Status', hue = 'Gender', data = df)
```

# **Q7. Showing City Category on the basis of age**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'City_Category', hue = 'Age', data = df)
```

# **Q8. Showing age on the basis of City Category**

```python
sns.set(rc = {'figure.figsize' : (12, 6)})
sns.countplot(x = 'Age', hue = 'City_Category', data = df)
```

# **Q9. Showing city category on the basis of gender**

```python
sns.set(rc = {'figure.figsize' : (8, 8)})
sns.countplot(x = 'City_Category', hue = 'Gender', data = df)
```

# **Q10. No of purchases from each city**

```python
df.groupby('City_Category').size().plot(kind = 'pie', autopct = '%.2f%%', figsize = (8, 8))
```

# **Q11. Total amount spend on products from each city**

```python
df.groupby('City_Category')['Purchase'].sum().plot(kind = 'pie', autopct = '%.2f%%', figsize = (8, 8))
```

# **Q12. Avg amount spend on products from each city**

```python
df.groupby('City_Category')['Purchase'].mean().plot(kind = 'pie', autopct = '%.2f%%', figsize = (8, 8))
```
